```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np 
```

```python
image = img.imread('bmw.jpg') 
plt.figure(figsize=(12,10))
plt.axis('off')
plt.imshow(image)
```

## ***NEGATIVE IMAGE***

```python
img_bgr = cv2.imread('bmw.jpg')

color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
    histr = cv2.calcHist([img_bgr], [i], None, [256], [0, 256]) 
    plt.plot(histr, color = col) 
    plt.xlim([0, 256]) 

plt.title('Histogram Plotting :Orginal Image')
plt.show() 
```

```python
height, width  , _ = img_bgr.shape
```

```python
for i in range (0, height - 1):
    for j in range (0, width - 1):

        #pixel value 
        pixel = img_bgr[i, j]

        #negating each channel 
        pixel [0] = 255 - pixel [0]
        pixel [1] = 255 - pixel [1]
        pixel [2] = 255 - pixel [2]

        #store new values
        img_bgr [i, j] = pixel



plt.imshow(img_bgr)
plt.show()
```

```python
color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
      
    histr = cv2.calcHist([img_bgr],  
                         [i], None, 
                         [256], 
                         [0, 256]) 
      
    plt.plot(histr, color = col) 
    plt.xlim([0, 256]) 
      
plt.title ('Histogram Plotting : Negative Image')
plt.show() 
```

## ***GRAY LEVEL SLICING***

```python
img_gray= cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
cv2.imwrite ('gray_bmw.jpg', img_gray)
```

```python
img1 = cv2.imread('gray_car.jpg')
cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Define the lower and upper thresholds for gray level slicing
low = 100
high = 200

img_sliced = np.where((img1 >= low) & (img1<= high), 255, 0).astype(np.uint8)
```

```python
fig , axes = plt.subplots(1,2, figsize=(16, 16))

axes [0].imshow(img1, cmap='gray')
axes [0].set_title('Original Image')

axes [1].imshow(img_sliced, cmap='gray')
axes [1].set_title('Sliced Image')

for ax in axes :
    ax.axis('off')

plt.show()
```

## ***HISTOGRAM EQUALIZATION***

```python
# Read the grayscale image
img1 = cv2.imread('gray_car.jpg', cv2.IMREAD_GRAYSCALE)
hist_eq = cv2.equalizeHist(img1)
```

```python
fig , axes = plt.subplots(1,2, figsize=(16, 16))

axes [0].imshow(img1, cmap='gray')
axes [0].set_title('Original Image')

axes [1].imshow(hist_eq, cmap='gray')
axes [1].set_title('Histogram Equalized')

for ax in axes :
    ax.axis('off')

plt.show()
```
